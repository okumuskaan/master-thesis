## 1. Graph Theory

### 1.1. Graph Definition

A graph is defined as:
$$ \mathcal{G} = \{ \mathcal{V}, \mathcal{E}, W \} $$

* $\mathcal{V}$: set of vertices with $|\mathcal{V}|=N$
* $\mathcal{E}$: set of edges with an edge $e = (i,j)$ connecting vertices $i$ and $j$
* $W$: weighted adjacency matrix with $W_{i,j}$ edge weight for an edge, o.w. it's equal to $0$

### 1.2. Graph Creation

To create a graph, a weighted adjacency matrix is enough to encode a graph. It's the most efficient representation to deal with when using spectral methods.

**Simple Graph:**

```python
W = np.array([
    [0,3,0,12,4],
    [3,0,5,0,0],
    [0,5,0,0,2],
    [12,0,0,0,7],
    [4,0,2,7,0]
])
G = pygsp.graphs.Graph(W)
myGraphPlot(G,  
            coords = [[0,5], [2,7], [5,6], [2,1], [5,3]],
            vertex_names=["A", "B", "C", "D", "E"],
            title = "A Simple Graph",
            display_vertexnames = True,
            display_edgeweights = True,
            print_graph_info=True)
```

**More Complex Graph:**

```python
W = np.random.uniform(size=(30, 30))  # Full graph.
W[W < 0.93] = 0  # Sparse graph.
W = W + W.T  # Symmetric graph.
np.fill_diagonal(W, 0)  # No self-loops.

G = pygsp.graphs.Graph(W)

myGraphPlot(G,
            title = "More Complex Graph")
```

**Generic Graph Construction Methods:** 

#### a.  Generation of Nodes with Node Coordinates:

A graph can be constructed from a set of points in $\mathbb{R}^d$.

#### b. Finding the Structure (Edges):

* Connect a point to its k-nearest neighbors - kNN graph
* Connect a point to points closer than some measure

#### c. Associate a weight to each edge:

- Gaussian Kernel Weighting function
$$
    W_{i,j} = \begin{cases}
        \exp (-\frac{[dist(i,j)]^2}{2\theta^2}), & \text{if} \: dist(i,j)\leq \mathcal{K} \\
        0, & \text{o.w.}
    \end{cases}
$$

or suppose that each edge is determined before:

$$
    W_{i,j} = \exp (-\frac{[dist(i,j)]^2}{2\theta^2})
$$

```python
n_points = 100
dimensionality = 2
points = np.random.uniform(size=(n_points, dimensionality))
plt.figure(figsize=(10,5))
plt.scatter(points[:, 0], points[:, 1]);
plt.title("Nodes and Their Location in $\mathbb{R}^2$")
plt.show()
```

```python
sigma=0.1

G = pygsp.graphs.NNGraph(points, NNtype='knn', k=2, sigma=sigma**2, rescale=False, symmetrize_type='maximum')
myGraphPlot(G,
            title = "kNN Graph with k=2 where each edge weight is constructed by Gaussian method")
```

## 1.3. Graph Properties



```python
G = pygsp.graphs.Logo()

print("Number of Nodes: \t", G.N)
print("Number of Edges: \t", G.Ne)

print("Is_connected: \t\t", G.is_connected())
print("Is_directed: \t\t", G.is_directed())
```

```python
print("Type of Weight matrix: ", type(G.W))

plt.spy(G.W, markersize=5)
plt.title("Sparsity Pattern of G.W")
plt.show()
```
