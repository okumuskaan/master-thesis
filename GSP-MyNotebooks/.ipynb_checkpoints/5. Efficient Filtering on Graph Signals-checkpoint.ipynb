## 5. Efficient Filtering on Graph Signals

### 4.1. Spectral Filtering

$$
f_{out} = g_\theta(L) f_{in}
$$

where $g_\theta(L) = Ug_\theta(\Lambda)U^T$ and $g_\theta(\Lambda) = \text{diag}(\theta)$. Here, $\theta_l = g(\lambda_l)$ and the function $g(.)$ represents the spectral function of the filter kernel.

* Non-localized in vertex domain
* Learning complexity in $O(n)$
* Computational complexity in $O(n^2)$


### 4.2. Polynomial Approximation of Filters

$$
g_\theta(\Lambda) = \sum_{k=0}^{K-1} \theta_k \Lambda^k
$$

where $\theta \in \mathbb{R}^K$ is a vector of polynomial coefficients.

By polynomial approximation, we can make filtering as a localized linear transform as:

$$
f_{out}(i) = \sum_{j \in \mathcal{N}(i,K) \cup \{i\}} b_{ij} f_{in}(j)
$$

$$
b_{ij} = (g_\theta(L) \delta_i)_j = (g_\theta(L))_{ij} = \sum_k \theta_k (L^k)_{ij}
$$

* Lemma: 
$$
d_{\mathcal{G}} (i,j) > K \:\:\: => \:\:\: (L^K)_{ij} = 0
$$

* Note that lemma can be proven from the fact that the element of power of adjacency matrix gives the number of K-hop paths from vertex i to vertex j.

Therefore, 
$$
b_{ij} = (g_\theta(L))_{ij} = \sum^K_{k=d_{\mathcal{G}} (i,j)} \theta_k (L^k)_{ij}$$

* K-localized operation
* Learning complexity in $O(K)$
* Cost is still $O(n^2)$ because of the multiplication with the Fourier basis

### 4.3. Recursive Formulation for Fast Filtering

Chebyshev approximation can be used to prevent spectral implementation:

$$
f_{out} = g_\theta(L) f_{in} = [\overline{f_{in, 0}},...,\overline{f_{in, K-1}}] \theta
$$

$$
\overline{f_{in, k}} = 2 \tilde{L} \overline{f_{in, k-1}} - \overline{f_{in, k-2}}
$$

with $\overline{f_{in, 0}}=f_{in}$ and $\overline{f_{in, 1}} = \tilde{L} f_{in}$ and $\tilde{L} = \frac{2L}{\lambda_{max}}-1$.

* Entire filtering operation costs $O(K|\mathcal{E}|)$, which is $<< O(n^2)$ for most of the case.

```python
G = pygsp.graphs.Logo()
G.compute_fourier_basis()

x = np.zeros(G.N)
x[G.info['idx_g']-1] = -1
x[G.info['idx_s']-1] = 0
x[G.info['idx_p']-1] = 1

rs = np.random.RandomState(42)
x_noisy = x + rs.uniform(-1, 1, size=G.N)

fig, axes = plt.subplots(1,2,figsize=(20,6))
myGraphPlotSignal(G, 
                s = x,
                title = "Clean Signal",
                ax = axes[0])
myGraphPlotSignal(G, 
                s = x_noisy,
                title = "Noisy Signal",
                ax = axes[1])

tau = 1
g = lambda x : 1/(1 + tau*x)
g = pygsp.filters.Filter(G, g)

x_denoised = g.filter(x_noisy, method='exact')
myGraphPlotSignal(G, 
                s = x_denoised,
                title = "Denoised Signal by Exact Filtering")
```

```python
def get_approx_filter(c):
    filt_approx = lambda x: np.sum(np.tile(c[1:], (len(x), 1)) * np.cos(np.tile(np.arange(len(c)-1) + 1, (len(x), 1)) * np.transpose(np.tile(np.arccos(x), (len(c) - 1, 1)))), 1) + np.tile(c[0], (len(x), 1)).T * np.cos(0 * np.arccos(x)) / 2
    return filt_approx
```

```python
G = pygsp.graphs.Logo()
G.compute_fourier_basis()

g = lambda x: 1/(1+x)
filt_g = pygsp.filters.Filter(G, g)
c = pygsp.filters.approximations.compute_cheby_coeff(filt_g, m=2)

filt_approx = get_approx_filter(c)
x = np.arange(0, G.lmax, (G.lmax) / 1000)
plt.figure()
plt.plot(x, g(x))
plt.plot(x, np.squeeze(filt_approx((x-G.lmax/2)/(G.lmax/2))))
plt.legend(['original filter', 'polynomial approximation'])
plt.show()
```

```python
start_time = time.time()
G.compute_fourier_basis(recompute=True)
x_denoised_exact = filt_g.filter(x_noisy, method='exact')
time_exact_filter = time.time() - start_time
start_time = time.time()
G.estimate_lmax(recompute = True)
x_denoised_cheby = filt_g.filter(x_noisy, method='chebyshev', order=2)
time_cheby_filter = time.time() - start_time
```

```python
fig, axes = plt.subplots(1,2,figsize=(20,6))
myGraphPlotSignal(G, 
                s = x_denoised_exact,
                title = "Denoised Signal by Exact Filtering",
                ax = axes[0])
myGraphPlotSignal(G, 
                s = x_denoised_cheby,
                title = "Denoised Signal by Chebyshev Approximation",
                ax = axes[1])

print('The max error is ' + str(np.max(np.abs(x_denoised_exact-x_denoised_cheby))))
print('The mean energy of the error is ' + str((np.sum((x_denoised_exact - x_denoised_cheby)**2)/G.N)))

print('The computation time for exact filtering was ' + str(time_exact_filter))
print('The computation time for approx filtering was ' + str(time_cheby_filter))
```

```python
G = pygsp.graphs.Logo()
G.compute_fourier_basis()

tau = 1
g = lambda x : 1/(1 + tau*x)
filt_g = pygsp.filters.Filter(G, g)

x_denoised_exact = filt_g.filter(x_noisy, method='exact')
```

```python
ms = np.arange(2,100,1)

max_errors = []
means_err_energy = []
times_cheby = []

for m in ms:
    start_time = time.time()
    G.estimate_lmax(recompute = True)
    x_denoised_cheby = filt_g.filter(x_noisy, method='chebyshev', order=m)
    time_cheby = time.time() - start_time
    
    max_errors.append(np.max(np.abs(x_denoised_exact-x_denoised_cheby)))
    means_err_energy.append((np.sum((x_denoised_exact - x_denoised_cheby)**2)/G.N))
    times_cheby.append(time_cheby)
    
fig, axes = plt.subplots(1,3,figsize=(20,6))
axes[0].plot(ms, max_errors)
axes[0].set_title("Max Error vs Order")
axes[1].plot(ms, means_err_energy)
axes[1].set_title("Mean of Error Energy vs Order")
axes[2].plot(ms, times_cheby)
axes[2].set_title("Time Elapsed vs Order")
plt.show()
```
